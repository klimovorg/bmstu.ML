## Анализируем по сахару в крови

### Изначальные данные (Данные - data)

```python
#fbs (уровень сахара в крови натощак > 120 мг / дл) (1 = true; 0 = false)
sns.jointplot(x='age', y='fbs', data=data, kind="kde")
```

### Смотрим тех, у кого уровень большой (Данные - data1)

```python
data1 = data[data['fbs'] == 1]
```

```python
print("Кол-во с большим уровнем сахара в крови:", data1.shape[0])
```

```python
print("________________________________________________")
print("Изначально было:")
print("мужчины", data['sex'].value_counts()[1])
print("женщины", data['sex'].value_counts()[0])
print("________________________________________________")
print("Считаем кол-во людей с привышенным сахором:")
print("мужчины", data1['sex'].value_counts()[1])
print("женщины", data1['sex'].value_counts()[0])
print("________________________________________________")
print("Процент людей с привышенным уровнем")
print("мужчины", data1['sex'].value_counts()[1] / data['sex'].value_counts()[1],"%")
print("женщины", data1['sex'].value_counts()[0] / data['sex'].value_counts()[0],"%")
```

```python
sns.boxplot(x=data1['age'])
```
